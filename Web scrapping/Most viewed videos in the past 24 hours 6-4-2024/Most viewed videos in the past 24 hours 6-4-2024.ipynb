# web scrapping 

## getting the data from the web page

the target web page is https://kworb.net/youtube/

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = r'https://kworb.net/youtube/'
page = requests.get(url)
soup = BeautifulSoup(page.text , 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')
```

## getting the needed table from the soup

id="youtuberealtime" , class="sortable"

```python
table = soup.find('table' , class_ = 'sortable')
print(table)
```

## getting table titles

```python
table_titles = soup.find_all('th')
table_titles = [title.text for title in table_titles]
```

```python
table_titles
```

```python
import pandas as pd

df = pd.DataFrame(columns= table_titles)
df
```

## getting table data

```python
table_data = table.find_all('tr')
```

```python
table_data
```

```python
for row in table_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    df.loc[len(df)] = individual_row_data
```

```python
df.head(10)
```

## Save CSV file

```python
df.to_csv('Most viewed videos in the past 24 hours.csv' , index=False)
#6/4/2024
```
